```python
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
out_path = r"G:\Araki\SMSigxSMAP\3_data_out\a_diff_202303\results_20230314.nc"
ds = xr.open_dataset(out_path)
```

```python
ds
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ds.a_diff.plot(ax=ax, cmap='RdYlBu_r')
ax.set_title('Difference in the dS/dt vs S slope')
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ds.MSE_upper.plot(ax=ax, vmin=-20)
ax.set_title('MSE upper quantile (Masked by SM availability)')
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ds.MSE_lower.plot(ax=ax, vmax=0, vmin=-20)
ax.set_title('MSE lower quantile (Masked by SM availability)')
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ds.PET_upper.where(ds.meanSM.notnull()).plot(ax=ax, vmax=6)
ax.set_title('PET upper quantile (Masked by SM availability)')
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ds.PET_lower.where(ds.meanSM.notnull()).plot(ax=ax, vmax=6)
ax.set_title('PET lower quantile (Masked by SM availability)')
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ds.meanSM.plot(ax=ax)
ax.set_title('Mean SM')
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ds.meanP.where(ds.meanSM.notnull()).plot(ax=ax)
ax.set_title('Mean P (Masked by SM availability)')
```

```python
PET_diff = ds.PET_upper-ds.PET_lower
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
PET_diff.where(ds.meanSM.notnull()).plot(ax=ax)
ax.set_title('Differences in PET (upper/lower quantiles) (Masked by SM availability)')
```

```python
a_diff_normalized = ds.a_diff/(ds.PET_upper-ds.PET_lower)
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
a_diff_normalized.plot(ax=ax, cmap='RdYlBu_r')
ax.set_title('a_diff normalized by PET')
```
